```python
def llstm():
    model = keras.Sequential()
    model.add(tf.keras.layers.LSTM(128, return_sequences=True, input_shape=[win, num_features]))
    model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
    model.add(keras.layers.BatchNormalization())
    model.add(tf.keras.layers.LSTM(128, return_sequences=True))
    model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
    model.add(keras.layers.BatchNormalization())
    model.add(tf.keras.layers.Dropout(0.3))
    model.add(tf.keras.layers.LSTM(64, return_sequences=True))
    model.add(layers.Dropout(0.3))
    model.add(tf.keras.layers.Dense(1))
    
    
    
    model.compile(loss="mae", optimizer=keras.optimizers.Adam(learning_rate=0.001), metrics=['mae'])
    return model
```

```python
llst = llstm()
llst.summary()
```

```python
early_stopping = keras.callbacks.EarlyStopping(monitor = 'val_loss',
                                               min_delta=0.00001,
                                               patience=100, 
                                               restore_best_weights=True)


EPOCHS = 1000

llst_history = llst.fit(
    train_generator,
    validation_data=val_generator, 
    epochs=EPOCHS, 
    callbacks=[early_stopping],
    batch_size = batch_siz,
    shuffle=False)
```

```python
import matplotlib.pyplot as mpl
mpl.plot(llst_history.history['loss'])
mpl.plot(llst_history.history['val_loss'])
mpl.xlabel('Epochs')
mpl.ylabel('Loss')
mpl.legend(['Train', 'Validation'])
mpl.show()
```

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

def tmodel():
    
  
    model = keras.Sequential()
### basic RNN model
    model.add(layers.SimpleRNN(units=128, return_sequences=True, input_shape=[win, num_features]))
    model.add(keras.layers.BatchNormalization())
    model.add(layers.Dropout(0.2))
    model.add(layers.SimpleRNN(units=128, return_sequences=True))
    model.add(keras.layers.BatchNormalization())
    model.add(layers.Dropout(0.2))
    model.add(layers.SimpleRNN(units=64, return_sequences=True))
    model.add(keras.layers.BatchNormalization())
    model.add(layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(1))


       
    
    model.compile(loss="mae", optimizer=keras.optimizers.Adam(learning_rate=0.001), metrics=['mae'])
    return model
```

```python
tmodel=tmodel()
tmodel.summary()
```

```python
early_stopping = keras.callbacks.EarlyStopping(monitor = 'val_mae',
                                               min_delta=0.0001,
                                               patience=100, 
                                               restore_best_weights=True)

# What it means is, monitor validation loss, if the change in loss is less than 0.0001 for 20 epochs, then stop training. 
# Additionally, it returns the best epoch weights

EPOCHS = 1000
#EPOCHS = 1
tmodel_history = tmodel.fit(
    train_generator,
    validation_data=val_generator, 
    epochs=EPOCHS, 
    callbacks=[early_stopping], 
    batch_size=batch_siz, 
    shuffle=False)
```

```python
import matplotlib.pyplot as mpl
mpl.plot(tmodel_history.history['loss'])
mpl.plot(tmodel_history.history['val_mae'])
mpl.xlabel('Epochs')
mpl.ylabel('Loss')
mpl.legend(['Train', 'Validation'])
mpl.show()
```

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)

def gruu():
    regressorGRU = keras.Sequential()
    regressorGRU.add(GRU(units=128, return_sequences=True, input_shape=(win, num_features)))
    regressorGRU.add(Dropout(0.2))
    regressorGRU.add(GRU(units=128, return_sequences=True, input_shape=(win, num_features)))
    regressorGRU.add(Dropout(0.2))
    regressorGRU.add(GRU(units=128, return_sequences=True, input_shape=(win, num_features)))
    regressorGRU.add(Dropout(0.2))
    regressorGRU.add(GRU(units=64))
    regressorGRU.add(Dropout(0.2))
    regressorGRU.add(Dense(units=1))
    regressorGRU.compile(loss="mae", optimizer=keras.optimizers.Adam(learning_rate=0.001), metrics=['mae'])


    return regressorGRU
    
```

```python
grumodel=gruu()
grumodel.summary()
```

```python
early_stopping = keras.callbacks.EarlyStopping(monitor = 'val_mae',
                                               min_delta=0.0001,
                                               patience=100, 
                                               restore_best_weights=True)

# What it means is, monitor validation loss, if the change in loss is less than 0.0001 for 20 epochs, then stop training. 
# Additionally, it returns the best epoch weights

EPOCHS = 1000
#EPOCHS = 1
grumodel_history = grumodel.fit(
    train_generator,
    validation_data=val_generator, 
    epochs=EPOCHS, 
    callbacks=[early_stopping], 
    batch_size=batch_siz, 
    shuffle=False)
```

```python
import matplotlib.pyplot as mpl
mpl.plot(grumodel_history.history['loss'])
mpl.plot(grumodel_history.history['val_mae'])
mpl.xlabel('Epochs')
mpl.ylabel('Loss')
mpl.legend(['Train', 'Validation'])
mpl.show()
```
